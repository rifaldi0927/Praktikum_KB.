```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Head

```python
df = pd.read_csv('Employee.csv')
df_iris = pd.read_csv('Employee.csv')
df_iris.head()
```

## Piechart

```python
data = df.Education.value_counts()
plt.pie(data, labels=data.index, autopct="%1.1f%%")
plt.title('Jumlah Data')
plt.show()
```

 ## Histogram

```python
plt.hist(df_iris['Education'], bins=15, color='skyblue')
plt.title('Education')
plt.ylabel('Frekuensi kemunculan')
plt.xlabel('Education')
plt.show()
```

## Scatter plot

```python
plt.scatter(df_iris['ExperienceInCurrentDomain'], df_iris['JoiningYear'])
plt.title('Ukuran Sepal')
plt.xlabel('Panjang sepal')
plt.ylabel('Lebar sepal')
plt.show()
```

## Sctter plot

```python
setosa = df_iris.loc[df_iris.Education == 'Bachelors']
versicolor = df_iris.loc[df_iris.Education == 'Masters']
virginica = df_iris.loc[df_iris.Education == 'PHD']
plt.scatter(setosa.City, setosa.Gender, c='green',
label='setosa')
plt.scatter(versicolor.City, versicolor.Gender, c='blue',
label='versicolor')
plt.scatter(virginica.City, virginica.Gender, c='red',
label='virginica')
```

## Heatmap

```python
import seaborn as sns

cor_matrix = df_iris.corr(numeric_only=True)
sns.heatmap(cor_matrix, cmap='coolwarm', annot=True, center=0)
plt.show()
```
